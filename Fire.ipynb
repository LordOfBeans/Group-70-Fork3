# Fire Incident in Pittsburgh Neighbourhoods

This section we will look at the number of fires that the fire department have to deal with in the past decade.
We will use official records from the Pittsburgh fire department.

The metric will rank the neighbourhood depending on how much incident happened in each of the neighbourhood. The more the merrier.

## Part 1: Importing data

Lets first import the data in, along side the python modules we will be using.

```python
import pandas as pd
import matplotlib.pyplot as plt
import geopandas
fire_data = pd.read_csv("Fire_Alarm.csv")
fire_data.head()
```

Ok, a bit too much information, with many information that we won't be needing. SO lets filter out the only data we need, which is the neighbourhood that the indicent happened in.

```python
fire_data[["neighborhood"]].head()
```

Thats much better.

## Part 2: Creating DataFrame
Now we can make a counter that tally up the total number of incidents that happened in each neighbourhood. To store this data, I'm using a **python dictionary**.

After getting the data tallied up and stored, we can use the dictionary to make a **pandas dataframe**, which will allow us to plot the data out.

```python
item_dictionary = {}
name = ""
counter = 0
neighbourhood_counter = 0
for index, row in fire_data.iterrows():
    item_name = row["neighborhood"]
    if (item_name not in item_dictionary.keys()):
        item_dictionary.update({item_name: [item_name, 1]})
        continue
    counter = item_dictionary[item_name]
    counter[1] += 1
    item_dictionary.update({item_name: counter})
neighborhood_df = pd.DataFrame.from_dict(item_dictionary, orient = "index")
neighborhood_df.columns = ["Neighbourhood", "Number of Incidents"]
neighborhood_df
```

You can see that there is a *neighbourhood* called NaN having a ton of indicidents. Sadly, this means a large part of these incidents happened ouside of all Pittsburgh neighbourhood. This data will be no use for us, so we will remove **NaN** from the dataset. 

While we are at it, we might as well sort the data out.

```python
#neighborhood_df = neighborhood_df.iloc[1:]
neighborhood_df.sort_values("Number of Incidents", ascending=[False])
```

Now all the data is sorted and the neighbourhoods are ranked. However, this does not help us visualize where in the city that these fires are happening. So lets plot the data out.

## Part 3: Plotting with Geopandas

```python
neighborhood = geopandas.read_file("Neighborhoods_.shp").set_index('hood')
fire_map = neighborhoods.merge(neighborhood_df, how='left', left_on='hood', right_on="Neighbourhood")
fire_map[['hood','Number of Incidents']].head()
```

```python
fire_map.plot(column='Number of Incidents')
```

```python
fire_map.plot(column='Number of Incidents',
              cmap='OrRd',            
              edgecolor="white",      
              legend=True,              
              legend_kwds={'label': "Number of Incidents"}, 
              figsize=(15, 10),         
              )
```
