# Best Pittsburgh Neighbourhood: Fire Incidents

This section we will look at the number of fires that the fire department have to deal with in the past decade.
We will use official records from the Pittsburgh fire department.

The metric will rank the neighbourhood depending on how much incident happened in each of the neighbourhood. The more the merrier.

## Part 1: Importing data

Lets first import the data in, along side the python modules we will be using.

```python
import pandas as pd
import matplotlib.pyplot as plt
import geopandas
fire_data = pd.read_csv("Fire_Alarm.csv")
fire_data.head()
```

Ok, a bit too much information, with many information that we won't be needing. Solets filter out the only data we need, which is the neighbourhood that the indicent happened in.

```python
fire_data[["neighborhood"]].head()
```

Thats much better.

## Part 2: Creating DataFrame
Now we can make a counter that tally up the total number of incidents that happened in each neighbourhood. To store this data, I'm using a **python dictionary**.

After getting the data tallied up and stored, we can use the dictionary to make a **pandas dataframe**, which will allow us to plot the data out.

```python
neighbourhood_dictionary = {}
name = ""
counter = 0
neighbourhood_counter = 0
for index, row in fire_data.iterrows():
    neighborhood = row["neighborhood"]
    time = row["alarm_time"]
    if(type(time) == str):
        if(time.startswith("2016") or time.startswith("2017") or time.startswith("2018") or time.startswith("2019") or time.startswith("2020")):
            if(neighborhood not in neighbourhood_dictionary.keys()):
                neighbourhood_dictionary.update({neighborhood: [neighborhood, 1]})
                continue
            counter = neighbourhood_dictionary[neighborhood]
            counter[1] += 1
            neighbourhood_dictionary.update({neighborhood: counter})
neighborhood_df = pd.DataFrame.from_dict(neighbourhood_dictionary, orient = "index")
neighborhood_df.columns = ["Neighbourhood", "Number of Incidents"]
neighborhood_df.dropna()
```

Not the most pretty thing, but it work well enough.

You can see that there is a *neighbourhood* called NaN having a ton of indicidents. Sadly, this means a large part of these incidents happened ouside of all Pittsburgh neighbourhood. This data will be no use for us, so we will remove **NaN** from the dataset. 

While we are at it, we might as well sort the data out.

```python
neighborhood_df = neighborhood_df.iloc[1:]
neighborhood_df.sort_values("Number of Incidents", ascending=[False])
```

Now all the data is sorted and the neighbourhoods are ranked. However, this does not help us visualize where in the city that these fires are happening. So lets plot the data out.

## Part 3: Plotting with Geopandas

For this part I will be using the python module called **geopandas** which allows me to plot data according to geographical locations.

I will first import the map of pittburgh into the file. Then I'll merge the fire incident data with the map so it can be viewed together.

Afterward I'll plot the map out to be viewed.

```python
neighborhood = geopandas.read_file("Neighborhoods_.shp").set_index('hood')
fire_map = neighborhood.merge(neighborhood_df, how='left', left_on='hood', right_on="Neighbourhood")
fire_map[['Neighbourhood','Number of Incidents']].head()
```

```python
fire_map.plot(column='Number of Incidents')
```

The map is rather unclear with the messy colouring. Lets reformat the graph so its clear which part of the city is "better".

```python
fire_map.plot(column='Number of Incidents',
              cmap='OrRd',            
              edgecolor="black",      
              legend=True,              
              legend_kwds={'label': "Number of Incidents"}, 
              figsize=(15, 10),         
              )
plt.axis("off")
```

Now it is clear to see which neighbourhood stoodout in the competition.

From the graph we can see some of the "best neighbourhood" are: Southside Flats, Central Business District, Squirrel Hill South, Carrick, and Hazelwood.

## Conclusion

Using the data provided, we can have a clear view on which neighbourhood is the best based on the rules. From the map we plotted out, we can also see that the "better" neighbourhood are somewhat spread out throughout the city, rather than concentrated in one section of the town. The northern part of the city is worse than the southern part since its number of incidents is considerably less than the southern part of the city.
